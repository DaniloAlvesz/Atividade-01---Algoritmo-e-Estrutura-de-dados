```python
quantidade_patinhos = int(input("Quantos patinhos foram passear? "))

for i in range(quantidade_patinhos,0,-1):
    print(f"{i} Patinhos foram passear, além das montanhas, para brincar. A mamãe gritou, quá, quá, quá, quá!")
    if i > 1:
        print("Mas, só", i - 1, "patinhos voltaram de lá.")
    else:
        print("E todos os patinhos voltaram de lá! :D")
```

```python
lista = [2,4,6,8,10]

print("1 elefante incomoda muita gente!")
for num in lista:
    print(f"{num} elefantes")
    for i in range(1, num + 1):
        print("incomodam")
    print("Muito mais!")
    prox = num + 1 
    print(f"{prox} elefantes incomodam muita gente!")
```
